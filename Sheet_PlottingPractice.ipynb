```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X = np.arange(10)
Y1 = X**2
Y2 = X**3
```

```python
plt.figure(figsize=(10,8))
plt.plot(X,Y1,label='Square',marker="*",color='brown')
plt.plot(X,Y2,label='Cube',marker=">")
plt.xlabel("Indexes")
plt.ylabel("Squares and Cubes")
plt.title("Plotting simple line graphs")
plt.gca().fill_between(X,Y1,Y2,alpha=0.25,facecolor='pink')
plt.legend()
plt.grid()
```

```python
company = ["Google","Amazon","Microsoft","Yahoo","Facebook"] 
revenue = [90,143,56,78,99] 
profits = [9,23,19,21,27] 
xpos = np.arange(len(company))
plt.figure(figsize=(10,9))
plt.xticks(xpos,company)
plt.ylabel("US Tech Stocks")
plt.bar(xpos-0.2,revenue,width=0.4,label="Revenue")
plt.bar(xpos+0.2,profits,width=0.4,label="Profits")
plt.legend()
```

```python
X = np.arange(10)
Y1 = X**2
Y2 = X**3
plt.figure(figsize=(10,8))
plt.plot(X,Y1,label="Squared",color='green',marker="^")
plt.plot(X,Y2,label="Cubed",color='blue',marker=">")
plt.xlabel('Indexes')
plt.ylabel('Squares and Cubes')
plt.gca().fill_between(X,Y1,Y2,facecolor='yellow',alpha=0.49)
plt.legend()
plt.grid()
```

```python
###Introducing the bar charts
company = ['Amazon','Yahoo','Google','Facebook','Twitter']
xpos = np.arange(len(company))
revenue = [107,75,156,122,80]
profits = [22,12,20,31,29]
plt.title('Company net profits and revenues(in dollars)')
plt.xticks(xpos,company)
plt.bar(xpos-0.2,revenue,width=0.4,label="Revenues",color='red')
plt.bar(xpos+0.2,profits,width=0.4,label="Profits",color='green')
plt.legend()
```

```python
## Using Scatter Plots
X = np.random.rand(25)
Y = np.random.rand(25)
z = np.sqrt(X**2 + Y**2)
plt.scatter(X,Y,marker = "*",c=z)
```

```python
### Using Histograms
X = 100 + 15*np.random.randn(1000)
plt.hist(X,bins=10,color="green",alpha=0.75)
```

```python
t1 = np.linspace(0,2*np.pi,200)
func = np.sin(t1**2) + np.cos(t1)
fig,ax = plt.subplots(1,2,sharey=True)
ax[0].plot(t1,t1**2)
ax[0].grid()
ax[1].plot(t1,func)
```
